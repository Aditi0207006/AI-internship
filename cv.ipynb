```python
import numpy as np
import matplotlib.pyplot as plt
image = np.full((16,16,3),100, dtype='uint8')
image[4:10,4:5] = [255,103,31]
image[4:10,5:6] = [255,255,255]
image[4:10,6:7] = [0,255,0]
cv2.imshow('image',image)
cv2.imwrite('image.jpg',image)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
